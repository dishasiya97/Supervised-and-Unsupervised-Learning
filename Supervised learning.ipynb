```python
#creating DataFrame
import os
from xml.etree import ElementTree
import pandas as pd
def f1():
    dfcolumns = ["itemid", "headline","text", "bip:topics", "dc.date.published", "XML filename"]
    dfrows = []
    path='C:\Data'

    files = []
    for r, d, f in os.walk(path):
        for file in f:
            if '.xml' in file:
                files.append(os.path.join(r,file))
        
    for name in files:
 
        bip = []       
        dom = ElementTree.parse(name)
     
        root = dom.getroot()
        itemno = root.get("itemid")
     
        heading = root.findtext("headline",default=None)
     
        for child in root:
   
            for n in child.iter('text'):
                for i in n:
                        texts = i.text
                                     
     
                for m in child.iter('metadata'):
           
                    for j in m.iterfind(".//codes[@class='bip:topics:1.0']"):
                       
                            for k in j.iter("code"):
                              
                              
                                bip.append(k.get('code'))
                                
                 
                    for p in m.iterfind(".//dc[@element='dc.date.published']"):
                               pdate=p.get('value')
        filename = itemno + "newsML.xml"
                           
     
  
        dfrows.append({ "itemid" : itemno, "headline" : heading, "text" : texts, 
                 "bip:topics" : bip, "dc.date.published" : pdate, "XML filename" : filename })
    dfoutput = pd.DataFrame(dfrows, columns = dfcolumns)
    print (dfoutput)
```

```python
import re
import nltk
import numpy as np

nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
def f3():
    stoplist = stopwords.words('english')
    col=['text']
    
    rows = []
    ps = PorterStemmer()
    lemmatizer = WordNetLemmatizer()
        
    for n in files:
            new = []
            dom = ElementTree.parse(n)
            r = dom.getroot()
       
            for cld in r:         
           
                  for n in cld.iter('text'):
                     for i in n:
                        m = i.text
                        o = re.sub('[\s,.]|[\s,"]|[\s,*]|[\s,\']|[\s,-]|[\s,%]|[\s,)]|[\s,/]|[\s,?]|\W\d{1,5}[^)(]*'," ", m)
                        p = ps.stem(o)
                      

                        q = lemmatizer.lemmatize(p)  
                        for word in q.split():
                           

                            if word not in stoplist:
                       
                       
                                new.append(word)
                                newtext=new
            rows.append({"text":newtext}) 
    df=pd.DataFrame(rows,columns=col)
    print(df)
```
